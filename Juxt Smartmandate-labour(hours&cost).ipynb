```python
plt.boxplot(df['lh'])
plt.title("Labour hours data including outliers")
plt.show()
```

```python

plt.title("labour cost data including outliers ")
plt.boxplot(df['lc'])
plt.show()
```

```python
#removing outliers in labour hours data
limit = df.lh.quantile(0.95)
valid_data = df[df['lh'] < limit]
```

```python
labour_hours_data=valid_data['lh']
labour_hours_data
```

```python
plt.boxplot(labour_hours_data)
plt.title("There are no outliers as we removed them in labour hours data")
plt.show()
```

```python
#removing outliers in labour cost data
limit = df.lc.quantile(0.95)
valid_dataa = df[df['lc'] < limit]
```

```python
labour_cost_data=valid_dataa['lc']
labour_cost_data
```

```python
plt.boxplot(labour_cost_data)
plt.title("There are no outliers as we removed them in labour cost data")
plt.show()
```

```python
mann=pd.DataFrame({'lh':labour_hours_data, 'lc':labour_cost_data.head(1540)
})
```

```python
mann
```

```python
#understanding Data using pair plot graph
import seaborn as sns
sns.pairplot(mann, kind="scatter")
plt.show()
```

```python
#correlation graph
plt.matshow(mann.corr())
plt.xticks(np.arange(2), ('lh', 'lc'))
plt.yticks(np.arange(2), ('lh', 'lc'))
plt.show()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula="lc~lh", data=mann).fit()
result.summary()
```

```python
#The Obtained Regression Equation is 
#Lc=-1.5379+74.0838(lh)
```

```python
pred=result.predict(mann)
pred
```

```python
df_c = pd.concat([df['lc'], pred], axis=1)
df_c.to_csv("Predicted.csv")
```
